```python
x = df.drop(columns=['T'])
y = df['T']
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
```

```python
encoder_labels = LabelEncoder()
encoded_y = encoder_labels.fit_transform(y)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train, y_test = train_test_split(x_scaled, encoded_y, random_state =42,test_size =0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential(
    [
        Dense(128, input_dim = x_train.shape[1], activation='relu'),
        Dense(64,activation='relu'),
        Dense(32,activation='relu'),
        Dense(26,activation='softmax')
    ]
)
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy' , metrics=['accuracy'])
```

```python
model.fit(x_train,y_train, epochs=20, batch_size=32, validation_split=0.2)
```

```python
predictions = model.predict(x_test)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'\n✅ Test Accuracy: {accuracy:.2f}')
```

```python
predicted_labels = encoder_labels.inverse_transform(np.argmax(predictions,axis=1))

actual_labels = encoder_labels.inverse_transform(y_test)
```

```python
# Print some predicted and actual results
print("Some Predicted and Actual Results:")
for i in range(10):
    print("Predicted:", predicted_labels[i], "Actual:", actual_labels[i])
```
